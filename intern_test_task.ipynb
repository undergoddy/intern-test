### Избавляемся от выбросов.

Удаляем значения в каждой фиче превышающие 3 сигмы вправо.

```python
no_cols = ['geo_h3_10', 'lat_10', 'lon_10', 'dt', 'pred']
for col in no_na_train.columns:
    if col in no_cols:
        continue
    else:
        std = (no_na_train[col].std() + 1) * 3 # Добавляем 1 чтобы не терять столбцы с категориальными значениями
        no_na_train = no_na_train.query("`{0}` < @std".format(col))
```

### Модель

```python
import shap
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
shap.initjs()
```

```python
cols = ['feat_89', 'feat_11', 'feat_8', 'feat_104', 'feat_67', 'feat_81', 'feat_10', 'feat_87', 'feat_3', 'feat_88']
```

```python
X = no_na_train.drop(columns={'geo_h3_10', 'lat_10', 'lon_10', 'dt', 'pred'})[cols]
y = no_na_train['pred']
```

```python
%%time
model = xgb.train(params = {"learning_rate": 0.01}, dtrain = xgb.DMatrix(X, label=y), num_boost_round = 500)
```

```python
explainer = shap.TreeExplainer(model)
```

```python
%%time
shap_values = explainer.shap_values(X)
```

```python
shap.force_plot(explainer.expected_value, shap_values[1,:], X.iloc[1,:])
```

```python
shap.force_plot(explainer.expected_value, shap_values[:100,:], X.iloc[:100,:])
```

```python
shap.summary_plot(shap_values, X, plot_size=(20,9))
```

Попробуем предсказать данные из всего начального датафрейма train.

```python
X_train = train[cols]
Y_true = train['pred']
```

```python
Y_pred = model.predict(xgb.DMatrix(X_train))
```

```python
print('RMSE -', round(mean_squared_error(Y_true, Y_pred, squared=False), 2))
print('MAPE -', round(mean_absolute_percentage_error(Y_true, Y_pred) * 100, 2), '%')
```

```python
pd.Series(Y_pred).describe()
```

```python
train.pred.describe()
```

```python
sns.distplot(pd.Series(Y_pred))
```

```python
sns.distplot(train.pred)
```

Модель сохранила общую форму распределения предсказываемой переменной.  
В среднем модель ошибается на 602.54 (RMSE), эта ошибка в среднем отличается от факта на 8.47% (MAPE).
